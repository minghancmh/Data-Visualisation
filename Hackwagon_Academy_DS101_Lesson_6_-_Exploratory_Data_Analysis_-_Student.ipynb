### How is `resale_price` related to `floor_area_sqm`?

<div class="alert alert-info"> <b>Hint</b>: 

1. Create 3 empty lists. (Third is optional).


2. Loop through each row in data. For each row:
    - append `floor_area_sqm` to the first list
    - append `resale_price` to the second list
    - (Optional) append `resale_price` / `floor_area_sqm` to the third list
    
    
3. Create a scatter plot `plt.scatter()`
    - the first two lists are used for your x & y values
    - (Optional) the third list is used for your colour (`c`) parameter</div>

````python
# Write your code below:
floor_area_sqm = []
resale_price = []
ratios = []
for row in data:
    resale = row['resale_price']
    area = row['floor_area_sqm']
    ratio = resale / area
    resale_price.append(resale)
    floor_area_sqm.append(area)
    ratios.append(ratio)
    
plt.figure(figsize = (12,6))
plt.scatter(floor_area_sqm,resale_price, c=ratios, cmap = 'plasma')
plt.colorbar(label = "price per sqm")
# plt.figure(figsize = (12,6))
````

````python
from collections import Counter
import mplleaflet
plt.figure(figsize=(16, 8))

# Step 1: Prepare data
#         Accumulate towns and their respective count into a dictionary
#         Accumulate towns and their respective lons into a dictionary
#         Accumulate towns and their respective lats into a dictionary
#         Then, extract out the values into three separate lists

counts = {}
longitudes = {}
latitudes = {}

# display(data[:2])
# START CODING HERE.
for row in data:
    town = row['town']
    long = row['town_lon']
    lat = row['town_lat']
    if town not in counts:
        counts[town] = 1
        longitudes[town] = long
        latitudes[town] = lat
    else:
        counts[town] += 1

# print(longitudes)
# # END CODE HERE.

x = []
y = []
z = []

for town, count in counts.items():
    x.append(longitudes[town])
    y.append(latitudes[town])
    z.append(count/10)  # <<< For size scaling
    
# print(z)
    
# # Step 2: Plot!
print(x)
print(y)
plt.scatter(x, y, c=z, s = z, alpha=0.5, cmap='Reds')
mplleaflet.display()
````
